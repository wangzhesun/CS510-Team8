```python
import pandas as pd
training = pd.read_csv('Training.csv')
training.head()
testing = pd.read_csv("Testing.csv")
testing.head()
filter_df = testing[testing['prognosis']=='AIDS']
# pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
filter_df.head()
```

```python
symptoms = training.columns
symptoms = symptoms[0:-1]
print(symptoms)
```

```python
training.describe()
```

```python
# split training and testing 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
X = training.iloc[:,:-1]
y = training.iloc[:,-1]
le = LabelEncoder()
le.fit(y)
Y = le.transform(y)
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KN = KNeighborsClassifier(n_neighbors=5)
KN.fit(X_train,y_train)
y_predict = KN.predict(X_test)
```

```python
from sklearn import metrics
accuracy = metrics.accuracy_score(y_test, y_predict)
print("Accuracy is: ",accuracy)
```

```python
import pickle
pickle.dump(KN,open('KN.pkl','wb'))
```

```python
KN = pickle.load(open('KN.pkl','rb'))
```

```python
sympton_list = training.columns
sympton_list = list(sympton_list.values)
print(sympton_list)
disease_list = le.classes_
print(disease_list)
```

```python
# prediction function
import numpy as np
def predict_disease(user_input_sym):
    input_vector = np.zeros(len(sympton_list)-1)
    for sym in user_input_sym:
        if sym in sympton_list:
            index = sympton_list.index(sym)
            input_vector[index] = 1
    pre_dis_index = KN.predict([input_vector])
    disease = disease_list[pre_dis_index]
    # print(pre_dis)
    
    return disease
```

```python
# testing example
user_input = input("Enter your symptoms")
user_symptoms = user_input.split(",")

print(user_symptoms)
predicted_disease = predict_disease(user_symptoms)
print(predicted_disease)
```
